#Import necessary libraries

```python
import torch
import torch.nn.functional as F #Contains various activation functions, loss functions, and other operations frequently used in neural networks.
from torchvision import datasets
from torchvision.transforms import v2 #Contains common image transformations to prepare data for neural networks.
```

#Load the dataset and Apply Transformations

```python
ds = datasets.FashionMNIST(
    root = 'data',
    train = True,
    download = True,
    transform = v2.Compose(
        [v2.ToImage(), v2.ToDtype(torch.float32, scale=True)]),
    target_transform=v2.Lambda(
        lambda y: F.one_hot(torch.tensor(y), num_classes=10).float()
    ),
)
```

The FashionMNIST dataset is loaded and configured to download if not present. The transformations are applied to both the images (converting them to float32 tensors) and the labels (converting them to a one-hot encoded format).
